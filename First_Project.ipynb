```python
import pandas as Data
```

```python
Provider = Data.read_csv("C:/Users/Shubham/Downloads/providers_data.csv")
Provider.head()
```

```python
Reciver = Data.read_csv("C:/Users/Shubham/Downloads/receivers_data.csv")
Reciver.head()
```

```python
Food_Listing = Data.read_csv("C:/Users/Shubham/Downloads/food_listings_data.csv")
Food_Listing.head()
```

```python
Claims_data = Data.read_csv("C:/Users/Shubham/Downloads/claims_data.csv")
Claims_data.head()
```

```python
# How many food providers and receivers are there in each city?
# Number of Food Provider in each City
Provider['City'].value_counts()
```

```python
# Number of Food Reciver in each City
Reciver['City'].value_counts()
```

```python
# Which type of food provider (restaurant, grocery store, etc.) contributes the most food?
Provider['Type'].value_counts()
```

```python
# What is the contact information of food providers in a specific city?
df = Data.DataFrame(Provider)
print(df[['City', 'Contact']])
```

```python
# Which receivers have claimed the most food?
Cn = Data.concat([Reciver, Claims_data], axis = 1)
DF = Data.DataFrame(Cn)
Sol = DF[['Receiver_ID', 'Food_ID']]
print(Sol)
```

```python
# Which receivers have claimed the most food?
DF[(DF['Food_ID'] >= 998) & (DF['Status'] == 'Completed')]
```

```python
# What is the total quantity of food available from all providers?
Food_Quan = Data.DataFrame(Food_Listing)
Food_Listing['Food_Name'].value_counts()
```

```python
Total = Food_Quan.pivot_table(index = ['Food_Name'], values = ['Quantity'], aggfunc = 'sum')
Total.sort_values(by = ['Quantity'], ascending = False)
```

```python
# What is the total quantity of food donated by each provider?
print('Total Quantity of Food :', Food_Listing['Quantity'].sum())
```

```python
# Which city has the highest number of food listings?
City = Food_Quan.pivot_table(index = ['Location'], values = ['Quantity'], aggfunc = 'sum')
City.sort_values(by = ['Quantity'], ascending = False)
```

```python
# What are the most commonly available food types?
Food_Types = Food_Quan.pivot_table(index = ['Food_Type', 'Food_Name'], values = ['Quantity'], aggfunc = 'sum')
Food_Types.sort_values(by = 'Food_Name', ascending = False)
```

```python
# Which provider has had the highest number of successful food claims?
Con = Data.concat([Food_Listing, Claims_data], axis = 1)
DF = Data.DataFrame(Con)
Sol = DF[['Provider_Type', 'Status', 'Claim_ID']]
print(Sol)
```

```python
Total = DF.pivot_table(index = ['Provider_Type', 'Status'], values = ['Claim_ID'], aggfunc = 'sum')
Total.sort_values(by = ['Claim_ID'], ascending = False)
```

```python
# What percentage of food claims are completed vs. pending vs. canceled?
print(Claims_data['Status'].value_counts(normalize = True))
```

```python
# What is the average quantity of food claimed per receiver?
print('Average Quantity of Food claimed per Reciver :', Claims_data['Claim_ID'].mean())
```
